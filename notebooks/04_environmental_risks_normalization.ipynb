# Environmental Risks Normalization

Environmental risks were collected as multivalued observations during household surveys.

This notebook converts textual risk descriptions into binary indicators for analytical use.

```python
# =============================================================================
# LOAD DATA
# =============================================================================

df = pd.read_csv("../data/processed/households_sample.csv")
```

```python
# =============================================================================
# SPLIT RISK TYPES
# =============================================================================

risks = (
    df[["id_household", "risk_types"]]
      .assign(
          risk_types=lambda x:
          x["risk_types"].str.split(";")
      )
      .explode("risk_types")
)
```

```python
# =============================================================================
# CREATE BINARY MATRIX
# =============================================================================

environmental_risks = pd.crosstab(
    risks["id_household"],
    risks["risk_types"]
)

environmental_risks = (
    environmental_risks
        .astype(int)
        .reset_index()
)

environmental_risks.head()
```

```python
# =============================================================================
# EXPORT
# =============================================================================

environmental_risks.to_csv(
    "../data/processed/environmental_risks_sample.csv",
    index=False
)
```
