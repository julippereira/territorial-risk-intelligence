# Social Programs Normalization

The original survey stored social program participation as a multivalued text field.

This notebook transforms the original field into a relational structure using binary indicators.

```python
# =============================================================================
# LOAD DATA
# =============================================================================

import pandas as pd

df = pd.read_csv("../data/processed/households_sample.csv")

df[["id_household", "social_programs"]].head()
```

```python
# =============================================================================
# SPLIT MULTI-VALUE FIELD
# =============================================================================

programs = (
    df[["id_household", "social_programs"]]
      .assign(
          social_programs=lambda x:
          x["social_programs"].str.split(";")
      )
      .explode("social_programs")
)

programs.head()
```

```python
# =============================================================================
# CREATE BINARY MATRIX
# =============================================================================

social_programs = pd.crosstab(
    programs["id_household"],
    programs["social_programs"]
)

social_programs = (
    social_programs
        .astype(int)
        .reset_index()
)

social_programs.head()
```

```python
# =============================================================================
# EXPORT
# =============================================================================

social_programs.to_csv(
    "../data/processed/social_programs_sample.csv",
    index=False
)
```
