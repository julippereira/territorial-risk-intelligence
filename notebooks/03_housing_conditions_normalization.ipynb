# Housing Conditions Normalization

Housing infrastructure attributes were originally stored as a single text field.

This notebook converts each condition into a binary variable.

```python
# =============================================================================
# LOAD DATA
# =============================================================================

df = pd.read_csv("../data/processed/households_sample.csv")
```

```python
# =============================================================================
# SPLIT CONDITIONS
# =============================================================================

housing = (
    df[["id_household", "housing_conditions"]]
      .assign(
          housing_conditions=lambda x:
          x["housing_conditions"].str.split(";")
      )
      .explode("housing_conditions")
)
```

```python
# =============================================================================
# CREATE BINARY MATRIX
# =============================================================================

housing_conditions = pd.crosstab(
    housing["id_household"],
    housing["housing_conditions"]
)

housing_conditions = (
    housing_conditions
        .astype(int)
        .reset_index()
)

housing_conditions.head()
```

```python
# =============================================================================
# EXPORT
# =============================================================================

housing_conditions.to_csv(
    "../data/processed/housing_conditions_sample.csv",
    index=False
)
```
